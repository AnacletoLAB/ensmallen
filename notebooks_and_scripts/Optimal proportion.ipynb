```python
def size(beta, p):
    return np.log(1 - beta) / np.log(1 - p)
```

```python
def time(beta, eta):
    return np.log(1 - beta) + beta * eta

def memory(beta, eta):
    return (1 - beta) * np.log(1 - beta) + beta * eta
```

# 3D plots

```python
from mpl_toolkits.mplot3d import Axes3D 
```

```python
betas = np.linspace(0.001, 1, 1000)
etas = np.linspace(1, 80, 20)
```

```python
values = [
        (
            [e] * len(betas),
        [
            memory(b, e)
            for b in betas
        ],
        [
            time(b, e)
            for b in betas
        ]
        )
        for e in etas
] 
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.gca(projection='3d')
for x, y, z in values:
    ax.scatter3D(x, y, z, color=cm.rainbow(np.linspace(0, 1, len(x) - 1) % 1))
ax.set_xlabel('eta')
ax.set_ylabel('memory')
ax.set_zlabel('time')
fig.show()
```

# 2D plots

```python
betas = np.linspace(0.001, 1, 1000)
etas = np.linspace(1, 80, 20)
```

```python
for e in etas:
    plt.scatter(
        [
            memory(b, e)
            for b in betas
        ],
        [
            time(b, e)
            for b in betas
        ],
        color=cm.rainbow(np.linspace(0, 1, len(betas) - 1))
    )
    plt.xlabel("memory")
    plt.ylabel("time")
    plt.title("Tradeoff for different betas with eta = %f"%e)
    plt.show()
```
